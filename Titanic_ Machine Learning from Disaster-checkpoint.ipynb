```python
from sklearn import tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_data = pd.read_csv('train.csv')
```

```python
titanic_data.head()
```

```python
# смотрим, где нет значений
titanic_data.isnull().sum()
```

```python
# формируем переменные для обучения, при этом удаляем неныжные столбцы
X = titanic_data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y = titanic_data.Survived
```

```python
# кодируем строковые переменные в числовые, т.к. решающие деревья не могут работать со строками
X = pd.get_dummies(X)
```

```python
X = X.fillna({'Age': X.Age.median()})
```

```python
X.isnull().sum()
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(X, y)
```

```python
# увеличить размер отображения решающего дерева
#from matplotlib.pylab import rcParams
#rcParams['figure.figsize'] = 200,150
```

```python
# визуализация решающих деревьев
tree.plot_tree(clf, class_names=['Negative', 'Positive'],
               filled=True)
```

```python
# получает датафрейм и разделяет на тренировочные и тестовые сабсеты
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
max_depth_values = range(1, 100)
```

```python
scores_data = pd.DataFrame()
```

```python
# формируем новый датасет для отображения score для разных значений глубин
for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
    temp_score_data = pd.DataFrame({'max_depth': [max_depth],
                                    'train_score': [train_score],
                                    'test_score': [test_score]})
    scores_data = scores_data.append(temp_score_data)
```

```python
scores_data.head()
```

```python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'], value_vars=['train_score', 'test_score'],
                          var_name='set_type', value_name='score')
```

```python
scores_data_long.head()
```

```python
sns.lineplot(x="max_depth", y="score", hue="set_type",
                  data=scores_data_long)
```

```python
scores_data = pd.DataFrame()
```

```python
for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
    mean_cross_val_score = cross_val_score(clf, X_train, y_train , cv=5).mean()
    
    temp_score_data = pd.DataFrame({'max_depth': [max_depth],
                                    'train_score': [train_score],
                                    'test_score': [test_score],
                                    'cross_val_score': [mean_cross_val_score]})
    scores_data = scores_data.append(temp_score_data)
```

```python
scores_data_long.head()
```

```python
sns.lineplot(x="max_depth", y="score", hue="set_type", data=scores_data_long)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = tree.DecisionTreeClassifier()
clf
```

```python
parametrs = {'criterion': ['gini', 'entropy'], 'max_depth': range(1,30)}
```

```python
grid_search_cv_clf = GridSearchCV(clf, parametrs, cv=5)
```

```python
grid_search_cv_clf.fit(X_train, y_train)
```

```python
grid_search_cv_clf.best_params_
```

```python
best_clf = grid_search_cv_clf.best_estimator_
```

```python
best_clf.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
y_pred = best_clf.predict(X_test)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
y_predicted_prob = best_clf.predict_proba(X_test)
```

```python
pd.Series(y_predicted_prob[:, 1]).hist()
```

```python
y_pred = np.where(y_predicted_prob[:, 1] > 0.9, 1, 0)
```

```python
pd.Series(y_predicted_prob[:, 1]).unique()
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_predicted_prob[:,1])
roc_auc= auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
